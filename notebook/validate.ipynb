```python
import numpy as np
from scipy import stats
rng = np.random.default_rng()
sample1 = stats.uniform.rvs(size=100, random_state=rng)
sample2 = stats.norm.rvs(size=110, random_state=rng)
response = stats.ks_2samp(sample1, sample2)
```

```python
round(response.pvalue,16)
#p-value above than 0.05 then it means the two data belong to the same distribution

# here in this example the data sets are not from same distributuion as p value is less than 0.05. 
# This means that the model needs to be retrained

# Reason for the value less than 0.05 is because the distribution is uniform in one and norm in anpther as below
#   sample1 = stats.uniform.rvs(size=100, random_state=rng)
#   sample2 = stats.norm.rvs(size=110, random_state=rng)"""
```
